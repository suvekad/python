```python
q1=data['charges'].quantile(0.25)
```

```python
q1
```

```python
q3=data['charges'].quantile(0.75)
```

```python
q3
```

```python
IQR=q3-q1
IQR
```

```python
low=q1-1.5*IQR
low
```

```python
up=q3+1.5*IQR
up
```

```python
capping_range=data.copy()
```

```python
capping_range['charges']=np.where(capping_range['charges'] < low,low,
          np.where(capping_range['charges'] > up,up,
                   capping_range['charges'])
                  )
```

```python
sns.kdeplot(capping_range['charges'],color='r')
```

```python
capping_median=data.copy()
capping_median
```

```python
median=capping_median['charges'].median()
median
```

```python
capping_median['charges']=np.where(capping_median['charges'] < low,median,
          np.where(capping_median['charges'] > up,median,
                   capping_median['charges'])
                  )
```

```python
sns.kdeplot(capping_median['charges'],color='r')
```
